### Step 1. Import the necessary libraries

```python
%matplotlib inline

import pandas as pd # Librería pandas
import numpy as np # Libraría numpy
```

### Step 2. Import the dataset from this US_Baby_Names_right.csv 

### Step 3. Assign it to a variable called baby_names.

```python
baby_names = pd.read_csv('data/US_Baby_Names_right.csv', sep=',')
```

### Step 4. See the first 10 entries

```python
baby_names.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
column_to_drop = "Unnamed: 0"
baby_names = baby_names.drop(column_to_drop, axis = 1)
column_to_drop = "Id"
baby_names = baby_names.drop(column_to_drop, axis = 1)
baby_names.head(10)
```

### Step 6. Is there more male or female names in the dataset?

```python
baby_names[['Gender']].describe()
```

Esto nos dice que hay más nombres femeninos que masculinos. Sin embargo, hay que tener en cuenta que la mayoría de los nombres se repiten. Así pues, habría que agrupar por género y luego obtener una descripción de los nombres, mirando la variable unique, que nos dice cuántos F o M hay sin tener en cuenta los repetidos:

```python
gender_group = baby_names.groupby('Gender')
gender_group[['Name']].describe()
```

Se puede ver que hay más nombres de bebes de sexo femenino que masculino.

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names.groupby('Name')
names.describe()
```

### Step 8. How many different names exist in the dataset?

Podemos aprovechar que en la agrupación anterior hemos agrupado todos los nombres. Por lo tanto, contando solamente el número de filas que tiene (que de hecho se ve justo arriba) podremos saber cuántos nombres diferentes hay.

```python
len(names)
```

### Step 9. What is the name with most occurrences?

```python
names.sum()[names.sum()['Count']==max(names.sum()['Count'])][['Count']]
```

### Step 10. How many different names have the least occurrences?

En primer lugar, la menor repetición de nombres que uno se puede encontrar es:

```python
min(names.sum()['Count'])
```

Por lo tanto, de nombres que sólo se repitan cinco veces hay:

```python
len(names.sum()[names.sum()['Count']==min(names.sum()['Count'])][['Count']])
```

Que representan un porcentaje del total de:

```python
len(names.sum()[names.sum()['Count']==min(names.sum()['Count'])][['Count']])/len(names)*100
```

### Step 11. What is the median name occurrence?

```python
names.sum()[names.sum()['Count']==np.median(names.sum()[['Count']])]
```

### Step 12. What is the standard deviation of names?

```python
names.std()[['Count']].head()
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
names.describe()[['Count']].head()
```
